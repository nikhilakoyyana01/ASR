```python
from tqdm import tqdm
import os
import numpy as np
def feature_extractior(file_name):
    data,sample_rate=librosa.load(file_name)
    mfccs_features=librosa.feature.mfcc(y=data,sr=sample_rate,n_mfcc=40)
    mfccs_scaled_features=np.mean(mfccs_features.T,axis=0)
    return mfccs_scaled_features

features=[]
c=0
i=0
for files in os.listdir("free-spoken-digit-dataset-master/recordings"):
    file_name="free-spoken-digit-dataset-master/recordings/"+files
    data=feature_extractior(file_name)
    class_label='Number_'+str(i)
    features.append([data,class_label])
    c=c+1
    if(c==300):
        c=0
        i=i+1
extracted_features=pd.DataFrame(features,columns=['features','class'])    
extracted_features.to_csv('features.csv',index=False)
extracted_features.head()
```

```python
x=np.array(extracted_features['features'].tolist())
y=np.array(extracted_features['class'].tolist())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pickle
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
with open('decision_tree.pkl','wb') as model_file:
    pickle.dump(clf,model_file)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred1=knn.predict(x_test)
accuracy1=accuracy_score(y_test,y_pred1)
print(accuracy1)
with open('K_Neighbour.pkl','wb') as model_file:
    pickle.dump(knn,model_file)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=3)
rf.fit(x_train,y_train)
y_pred2=rf.predict(x_test)
accuracy2=accuracy_score(y_test,y_pred2)
print(accuracy2)
with open('random_forest.pkl','wb') as model_file:
    pickle.dump(rf,model_file)
```
